# 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as -
• Jan - Mar : Q1
• Apr - Jun : Q2
• Jul - Sep : Q3
• Oct - Dec : Q4

```python
print(sales.Month.unique())
```

```python
sales["Qtr"] = np.where((sales.Month =='Jan')|(sales.Month == 'Feb')|(sales.Month == 'Mar'), "Q1",
               np.where((sales.Month =='Apr')|(sales.Month == 'May')|(sales.Month == 'Jun'), "Q2",
               np.where((sales.Month =='Jul')|(sales.Month == 'Aug')|(sales.Month == 'Sep'), "Q3",
               np.where((sales.Month =='Oct')|(sales.Month == 'Nov')|(sales.Month == 'Dec'), "Q4","NaN"))))
                   
sales.head()
```

Compare Qtr wise sales in 2015 and 2016 in a bar plot

```python
sales7= sales.groupby(['Qtr']).aggregate({'Sales2015':'sum','Sales2016':'sum'})
sales7= sales7.round(decimals=0)
print(sales7)
sales7graph=sales7.plot.bar(figsize=(12,6))
```

```python
sales.Tier.unique()
```

```python
sales8 = sales[sales['Tier']!='Out']
sales8 = sales8.groupby(['Qtr','Tier']).aggregate({'Sales2016':'sum'})
sales8 = sales8.round(decimals=0)
sales8 = sales8.stack() # long to wide data
sales8 = sales8.unstack(level=0)
sales8
```

```python
sales8.plot(kind='pie', 
            labels=['High', 'Low', 'Medium'], 
            colors=['deepskyblue', 'darkorange', 'darkgreen'],
            autopct='%1.0f%%',
            fontsize=10,
            subplots='true',
            layout=(2,2), #to get a grid
            figsize=(20,20))
plt.show()
```
